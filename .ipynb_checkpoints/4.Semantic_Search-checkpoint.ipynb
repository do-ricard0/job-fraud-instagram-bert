### Import Data & Library 

```python
import pandas as pd
import spacy
import re
```

```python
df=pd.read_csv('labeled.csv')
```

```python
df.head()
```

### Extraction Company Name 

```python
nlp = spacy.load("en_core_web_trf")
```

```python
def extract_company_from_email_or_text(text):
    # Mencari alamat email dalam teks menggunakan ekspresi reguler
    emails = re.findall(r'\b[\w.-]+@[\w.-]+\.\w+\b', text)
    
    # Cek setiap email yang valid
    for email in emails:
        # Gunakan spaCy untuk mengekstrak nama perusahaan dari email
        doc = nlp(email)
        company_names = [ent.text for ent in doc.ents if ent.label_ == "ORG"]
        if company_names:
            return ", ".join(company_names)

    # Jika tidak ada nama perusahaan yang ditemukan dalam email,
    # lakukan ekstraksi nama perusahaan dari teks OCR
    doc = nlp(text)
    company_names = [ent.text for ent in doc.ents if ent.label_ == "ORG"]
    if company_names:
        return ", ".join(company_names)

    # Jika spaCy tidak menemukan entitas perusahaan, ekstrak dari nama email
    if emails:
        # Mengambil nama perusahaan dari bagian sebelum '@' dari email pertama
        return emails[0].split('@')[0].capitalize()

    return ""
```

```python
df['Extracted_Companies'] = df['OCR_Text'].apply(extract_company_from_email_or_text)
```

```python
df.to_csv('semantic_search_.csv',index=False)
```

```python
df.to_excel('semantic_search.xlsx', index=False)
```
